```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_v_line(x):
    if x is not None:
        global line
        line.remove()
        line = plt.axvline(x)
        plt.draw()
```

```python
fig = plt.gcf()
axes = plt.gca()
points = np.random.randn(500, 2)
plt.scatter(points[:, 0], points[:, 1])
line = plt.axvline(0)
```

```python
fig.canvas.mpl_connect("motion_notify_event", lambda x: plot_v_line(x.xdata))
```

```python
plt.show()
```
